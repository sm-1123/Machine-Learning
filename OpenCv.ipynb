```python
import cv2

import matplotlib.pyplot as plt

image = cv2.imread(r'C:\Users\Sourav Mukherjee\OneDrive\Pictures\army.jpg')

#checking if the image was loaded successfully

if image is not None:
    
    print("Image Dimensions:", image.shape)
    
    #converting BGR image to Rgb
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    plt.imshow(image_rgb)
    
    plt.axis('off') #hide axes
    
    plt.show()
    
else:
    
    print("Error loading image")
```

Reading image as greyscale using imread() Method

```python
import cv2

import matplotlib.pyplot as plt

img = cv2.imread(r'C:\Users\Sourav Mukherjee\OneDrive\Pictures\army.jpg', cv2.IMREAD_GRAYSCALE)

print('image dimension:', img.shape)

#Displaying image

plt.imshow(img, cmap='gray')

plt.axis('off')

plt.show()
```

Reading image with transparency channel using imread() Method

```python
import cv2

import matplotlib.pyplot as plt

img = cv2.imread(r'C:\Users\Sourav Mukherjee\OneDrive\Pictures\army.jpg', cv2.IMREAD_UNCHANGED)

print('image dimension:', img.shape)

if img.shape[2] == 4:
    
    img_rgba = cv2.cvtColor(img, cv2.COLOR_BGRA2RGBA)
    
else:
    
    img_rgba = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Displaying image

plt.imshow(img_rgba)

plt.axis('off')

plt.show()
```

OpenCV cv2.imread() and Color Channels

```python
#get blue channel
b1 = img[:,:,0]

#get green channel
g1 = img[:,:,1]

#get red channel
r1 = img[:,:,2]

#display the channels
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(b1, cmap='gray')
axes[0].set_title('Blue Channel')
axes[0].axis('off')
    
axes[1].imshow(g1, cmap='gray')
axes[1].set_title('Green Channel')
axes[1].axis('off')
    
axes[2].imshow(r1, cmap='gray')
axes[2].set_title('Red Channel')
axes[2].axis('off')

plt.show()
```
